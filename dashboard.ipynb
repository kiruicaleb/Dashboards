```python
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv("Grocery_sales.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

Introduction of visualization using dashboard to get a figurative representation of the relationship between sales and profits using the plotly packages.

```python
fig = make_subplots(rows=2, cols=1, subplot_titles=("Sales", "Profit"))
fig.add_trace(go.Scatter(x=df['Category'], y=df.groupby('Category')['Sales'].sum(), mode='lines', name='Sales'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['Category'], y=df.groupby('Category')['Profit'].sum(), mode='lines', name='Profit'), row=2, col=1)
fig.show()
```

```python
dash = make_subplots(rows=2, cols=1, subplot_titles=("Sales", "Profit"))
dash.add_trace(go.Scatter(x=df['Category'], y=df['Sales'], mode='lines', name='Sales'), row=1, col=1)
dash.add_trace(go.Scatter(x=df['Category'], y=df['Profit'], mode='lines', name='Profit'), row=2, col=1)
dash.show()
```

```python
fig = make_subplots(rows=2, cols=1, subplot_titles=("Sales", "Profit"))
fig.add_trace(go.Scatter(x=df['Category'], y=df.groupby('Region')['Sales'].sum(), mode='lines', name='Sales'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['Category'], y=df.groupby('Region')['Profit'].sum(), mode='lines', name='Profit'), row=2, col=1)
fig.show()
```

From the observations above below and above, it is evident that the dashboards communicate a common trend in sales and profits. It is clear that in the subsets of regions and categories, sales are generally proportional to profits hence a derivation that more sales are likely to increase margin of profits.

```python
fig = make_subplots(rows=2, cols=1, subplot_titles=("sales", "profits"))
fig.add_trace(go.Scatter(x=df['Region'], y=df.groupby('Category')['Sales'].sum(), mode='lines', name='Regional sales'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['Region'], y=df.groupby('Category')['Profit'].sum(), mode='lines', name='Regional profits'), row=2, col=1)
fig.show()
```
